```python
from sklearn.preprocessing import LabelEncoder


# Initialize the LabelEncoder
le = LabelEncoder()

# Fit and transform the data
train['Personality'] = le.fit_transform(train['Personality'])

# Show the results
train['Personality']
```

```python
X = train.drop(['Response Id', 'Personality'], axis = 1)
y = train['Personality']
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )
knn_model.fit(X_train, y_train)
KNeighborsClassifier()
```

```python
y_knn_pred= knn_model.predict(X_test)
y_knn_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score
knn_cr=classification_report(y_test, y_knn_pred)
print(knn_cr)
```

# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)
GaussianNB()
```

```python
y_gnb_pred= gnb_model.predict(X_test)
y_gnb_pred
```

```python
gnb_cr=classification_report(y_test, y_gnb_pred)
print(gnb_cr)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_model= DecisionTreeClassifier(criterion='entropy', random_state=5)
dt_model.fit(X_train, y_train)
DecisionTreeClassifier(criterion='entropy', random_state=5)
```

```python
y_dt_pred= dt_model.predict(X_test)
y_dt_pred
```

```python
dt_cr=classification_report(y_test, y_dt_pred)
print(dt_cr)
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
RandomForestClassifier()
```

```python
y_rf_pred= rf_model.predict(X_test)
y_rf_pred
```

```python
rf_cr=classification_report(y_test, y_rf_pred)
print(rf_cr)
```

# Conclusion

Here, KNN model gives the highest accuracy,followed by random forest, naive bayes and decision tree algorithm.
